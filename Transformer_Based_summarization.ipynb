### BART Transformer

```python
from transformers import BartTokenizer, BartForConditionalGeneration

def multi_doc_summarization(doc_list):
    # Load pre-trained model and tokenizer
    model = BartForConditionalGeneration.from_pretrained('facebook/bart-large-cnn')
    tokenizer = BartTokenizer.from_pretrained('facebook/bart-large-cnn')

    # Concatenate all documents
    combined_text = ' '.join(doc_list)

    # Tokenize text
    inputs = tokenizer([combined_text], max_length=1024, return_tensors='pt')

    # Generate summary
    summary_ids = model.generate(inputs['input_ids'], num_beams=4, max_length=5, early_stopping=True)
    summary = [tokenizer.decode(g, skip_special_tokens=True, clean_up_tokenization_spaces=False) for g in summary_ids]

    return summary[0]

# Example usage:
documents = ["This is document 1...", "This is document 2...", "This is document 3..."]
print(multi_doc_summarization(documents))
```

```python
from transformers import BartTokenizer, BartForConditionalGeneration

def multi_doc_summarization(doc_list):
    # Load pre-trained model and tokenizer
    model = BartForConditionalGeneration.from_pretrained('facebook/bart-large-cnn')
    tokenizer = BartTokenizer.from_pretrained('facebook/bart-large-cnn')

    # Concatenate all documents
    combined_text = ' '.join(doc_list)

    # Tokenize text
    inputs = tokenizer([combined_text], max_length=50, return_tensors='pt')

    # Generate summary
    summary_ids = model.generate(inputs['input_ids'], num_beams=4, max_length=100, early_stopping=True)
    summary = [tokenizer.decode(g, skip_special_tokens=True, clean_up_tokenization_spaces=False) for g in summary_ids]

    return summary

# Example usage:
# documents = ["This is document 1...", "This is document 2...", "This is document 3..."]
documents = all_text_documents[0]
summary = multi_doc_summarization(documents)
```
